# 1 Introducción

**Aprendizaje profundo:** $\quad$ área del ML especializada en el aprendizaje de redes con (muchas) capas ocultas

**Revolución del aprendizaje profundo:** $\quad$ el aprendizaje profundo alcanza gran popularidad a partir de los 2010, principalmente por la mejora de resultados en reconocimiento automático del habla (2011) y clasificación de imágenes (2012)

**Objetivo:** $\quad$ introducir arquitecturas destacadas del aprendizaje profundo para clasificación de imágenes principalmente

**LeNet (1998):** $\quad$ red **convolucional (CNN)** pionera, introducida en 1998 por Yann LeCun para MNIST
* Propone el uso de capas convolucionales y pooling (agrupación)

**AlexNet (2012):** $\quad$ CNN de Alex Krizhevsky para ImageNet que atrajo mucho interés
* Resultados espectaculares gracias a la profundidad, uso de GPUs, ReLUs y dropout

**Inception (2014):** $\quad$ CNN de Google que introduce la meta-arquitectura típica de las CNNs modernas

**ResNet (2015):** $\quad$ CNN de Microsoft que introduce **conexiones residuales** para entrenar modelos muy profundos
* Propone el uso de conexiones residuales y añade capas de normalización

**Transformer (2017):** $\quad$ arquitectura de Google para traducción automática basada en el **mecanismo de atención**
* Se considera la base de los modelos IA actuales más competitivos

**Vision Transformer (2020):** $\quad$ adaptación de Transformer para visión
