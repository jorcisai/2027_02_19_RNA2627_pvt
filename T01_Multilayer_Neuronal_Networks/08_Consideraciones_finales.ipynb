# 8 Consideraciones finales

**Antecedentes de las redes:** $\;$ las ideas básicas tras las redes son conocidas durante varias décadas (desde la cibernética)

**Conexiones con la biología:** $\;$ muchos autores se han inspirado en la biología

**Gran popularidad a partir de los 2010:** $\;$ por la mejora de resultados en reconocimiento automático del habla (2011) y clasificación de imágenes (2012)

**GPUs y grandes conjuntos de datos supervisados:** $\;$ son los principales factores tras la mejora de resultados

**Librerías de código abierto:** $\;$ facilitan la aplicación de las redes a multitud de problemas
* **Tensorflow (Google):** $\;$ [https://www.tensorflow.org/](https://www.tensorflow.org/)
* **PyTorch (Meta):** $\;$ [https://pytorch.org/](https://pytorch.org/)

**No necesidad de más de una capa oculta:** $\;$ un MLP de **una sola capa oculta** con suficientes unidades es una **aproximador de funciones universal** ya que puede aproximar toda función (mínimamente suave) tanto como se quiera

**Importancia de la profundidad:** $\;$ las redes profundas suelen ofrecer mejores resultados ya que las últimas capas aprovechan las características aprendidas en las primeras capas
