# 3 Multilayer Perceptrons

**Multilayer perceptron (MLP):** $\;$ composition of $\,L\,$ functions or **layers** $\,f^{(\ell)}(\boldsymbol{x}^{(\ell)};\boldsymbol{\theta}^{(\ell)})$
$$f(\boldsymbol{x};\boldsymbol{\theta})=f^{(L)}(f^{(L-1)}(\cdots f^{(1)}(\boldsymbol{x})\cdots))%
\quad\text{con}\quad%
\boldsymbol{x}^{(\ell+1)}=f^{(\ell)}(\boldsymbol{x}^{(\ell)})=\boldsymbol{\varphi}^{(\ell)}(\boldsymbol{a}^{(\ell)})
\quad\text{y}\quad%
\boldsymbol{a}^{(\ell)}=\mathbf{W}^{(\ell)}\boldsymbol{x}^{(\ell)}+\boldsymbol{b}^{(\ell)}$$

**Preactivation of the layer $\ell,\boldsymbol{a}_{\ell}$:** $\quad$ afinity of the previous layer output, $\,\boldsymbol{x}^{(\ell)};\,$ with $\,\boldsymbol{x}^{(1)}=\boldsymbol{x}$

**Activation/output of layer $\ell,\boldsymbol{x}^{(\ell+1)}$:** $\;$ basic transformation of $\,\boldsymbol{a}^{(\ell)}\,$ with $\,\varphi:\mathbb{R}\to\mathbb{R}\,$ **diferentiable** (and non-lineal)

**Usual terminology:**

* **Input layer:** $\quad\boldsymbol{x}^{(1)}=\boldsymbol{x}$

* **Hidden layers:** $\quad\boldsymbol{x}^{(\ell+1)}$, whose dimensionality indicates the number of outputs or **hidden units** of the layer

* **Capa de salida:** $\quad\boldsymbol{x}^{(L+1)},\,$ usually sigmoid or softmax activation for clasification; identity for regression

**Shallow networks (without hidden layers):** $\quad$ if $\,L=1,\,$ boiled down to conventional linear models

* **Binary logistic regression:** $\quad f(\boldsymbol{x};\boldsymbol{\theta})=x^{(2)},\;x^{(2)}=\sigma(a^{(1)}),\;a^{(1)}=\mathbf{W}^{(1)}\boldsymbol{x}+b^{(1)}$

* **Logistic regression:** $\quad f(\boldsymbol{x};\boldsymbol{\theta})=\boldsymbol{x}^{(2)},\;\boldsymbol{x}^{(2)}=\mathcal{S}(\boldsymbol{a}^{(1)}),\;\boldsymbol{a}^{(1)}=\mathbf{W}^{(1)}\boldsymbol{x}+\boldsymbol{b}^{(1)}$

* **Linear regression:** $\quad f(\boldsymbol{x};\boldsymbol{\theta})=x^{(2)},\;x^{(2)}=a^{(1)},\;a^{(1)}=\mathbf{W}^{(1)}\boldsymbol{x}+b^{(1)}$

**Deep neural networks (with hidden layers):** $\quad$ si $\,L>1,\,$ linear models with integrated **feature engineering**

**Solution:** $\quad$ MLP for XOR with tempered sigmoids

<div align="center"><img src="MLPXOR.svg" width="300"/></div>

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.special import expit
w1 = np.array([1, 1]); b1 = -1.5; w2 = np.array([1, 1]); b2 = -0.5; wy = np.array([-1, 1]); by = -0.5
P = lambda x, w, b: expit(10 * (w @ x + b)); MLPXOR = lambda x: P(np.array([P(x, w1, b1), P(x, w2, b2)]), wy, by)
fig, ax = plt.subplots(1, 1, figsize=(2.5, 2)); ax.grid()
ticks = [.0, 1.]; ax.set_xticks(ticks); ax.set_yticks(ticks)
R = np.linspace(-.05, 1.05, 50); x1, x2 = np.meshgrid(R, R)
XX = np.c_[np.ravel(x1), np.ravel(x2)]; yy = np.apply_along_axis(MLPXOR, 1, XX)
ax.contour(x1, x2, yy.reshape(x1.shape), 1, colors='orange', linestyles='solid')
cp = ax.contourf(x1, x2, yy.reshape(x1.shape), 1, cmap='Blues')
plt.colorbar(cp, ax=ax, shrink=0.8); # ax.scatter(*X.T, c=y, s=64);
```
