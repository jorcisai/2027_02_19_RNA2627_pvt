# 8 Final considerations

**Background of Networks:** The basic ideas behind networks have been known for several decades (since cybernetics).

**Connections with Biology:** Many authors have been inspired by biology.

**Greater Popularity Since the 2010s:** Due to improved results in automatic speech recognition (2011) and image classification (2012).

**GPUs and Large Supervised Datasets:** These are the main factors behind the improved results.

**Open Source Libraries:** These facilitate the application of networks to a multitude of problems.

* **TensorFlow (Google):** [https://www.tensorflow.org/](https://www.tensorflow.org/)

* **PyTorch (Meta):** [https://pytorch.org/](https://pytorch.org/)

**No Need for More Than One Hidden Layer:** A single-layer hidden-layer MLP with sufficient units is a universal function approximator because it can approximate any (minimally smooth) function as well as desired.

**Importance of depth**: Deeper networks generally offer better results because the later layers leverage the features learned in the earlier layers.
