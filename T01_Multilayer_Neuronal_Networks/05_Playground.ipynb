**URL y github:** $\quad$ [http://playground.tensorflow.org](http://playground.tensorflow.org) $\quad$ [https://github.com/tensorflow/playground](https://github.com/tensorflow/playground)

**Ejercicio:** $\quad$ for each classification task, search for the smallest network that solves the task


<p style="page-break-after:always;"></p>


**Solution:** $\quad$
* The linear form does not require hidden layers.
* The circle-ring form also does not require quadratic features; that, or a hidden layer of three units.
* The quadrant form is solved with a hidden layer of four units.
* The spiral form can be solved with quadratic features and two hidden layers of eight and six units.
