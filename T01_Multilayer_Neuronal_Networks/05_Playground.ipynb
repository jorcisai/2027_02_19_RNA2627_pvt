**Ejercicio:** $\quad$ por cada tarea de clasificación, halla una red minimalista que la resuelva


<p style="page-break-after:always;"></p>


**Solución:** $\quad$
* La lineal no requiere capas ocultas
* La círculo-anillo tampoco con características cuadráticas; eso, o una capa oculta de tres unidades
* La de cuadrantes se resuelve con una capa oculta de cuatro unidades
* La de espirales puede resolverse con características cuadráticas y dos capas ocultas de ocho y seis unidades
