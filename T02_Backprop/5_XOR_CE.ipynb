# 5 Example for XOR with softmax output

**Model:** $\quad$ hidden layer with two ReLUs and output layer with softmax for two classes
$$\begin{align*}
\boldsymbol{x}^{(1)}&\in\{0,1\}^2\\[3mm]
\boldsymbol{x}^{(2)}&=\mathbf{W}^{(1)}\boldsymbol{x}^{(1)}+\boldsymbol{b}^{(1)}\in\mathbb{R}^2\\[3mm]
\boldsymbol{x}^{(3)}&=\operatorname{ReLU}(\boldsymbol{x}^{(2)})\in\mathbb{R}^2\\[3mm]
\boldsymbol{x}^{(4)}&=\mathbf{W}^{(3)}\boldsymbol{x}^{(3)}+\boldsymbol{b}^{(3)}\in\mathbb{R}^2\\[3mm]
\boldsymbol{x}^{(5)}&=\hat{\boldsymbol{y}}=\mathcal{S}(\boldsymbol{x}^{(4)})\in[0,1]^2\\[3mm]
x^{(6)}&=\mathcal{L}=\mathbb{H}(\boldsymbol{y},\hat{\boldsymbol{y}})=-\boldsymbol{y}^t\log(\hat{\boldsymbol{y}})
\end{align*}$$

**Current parameters:**
$$\mathbf{W}^{(1)}=\begin{pmatrix}1&1\\1&1\end{pmatrix}\quad \boldsymbol{b}^{(1)}=\begin{pmatrix}-1\\0.5\end{pmatrix}\quad \mathbf{W}^{(3)}=\begin{pmatrix}1&-1\\-1&1\end{pmatrix}\quad\boldsymbol{b}^{(3)}=\begin{pmatrix}1\\-1\end{pmatrix}$$

**Objective:** $\;$ backprop with $\quad\mathbf{X}=\begin{pmatrix}\boldsymbol{x}_1^t\\\boldsymbol{x}_2^t\\\boldsymbol{x}_3^t\\\boldsymbol{x}_4^t\end{pmatrix}%
=\begin{pmatrix}0&0\\0&1\\1&0\\1&1\end{pmatrix}\quad$ y $\quad\mathbf{Y}=\begin{pmatrix}\boldsymbol{y}_1^t\\\boldsymbol{y}_2^t\\\boldsymbol{y}_3^t\\\boldsymbol{y}_4^t\end{pmatrix}=\begin{pmatrix}1&0\\0&1\\0&1\\1&0\end{pmatrix}$


<p style="page-break-after:always;"></p>


**Forward:**
$$\begin{align*}
\mathbf{X}^{(2)}&%
=\mathbf{X}^{(1)}\mathbf{W}^{(1)t}+\boldsymbol{1}\boldsymbol{b}^{(1)t}%
=\begin{pmatrix}0&0\\0&1\\1&0\\1&1\end{pmatrix}\begin{pmatrix}1&1\\1&1\end{pmatrix}+\begin{pmatrix}1\\1\\1\\1\end{pmatrix}(-1, 0.5)%
%=\begin{pmatrix}0&0\\1&1\\1&1\\2&2\end{pmatrix}+\begin{pmatrix}-1&0.5\\-1&0.5\\-1&0.5\\-1&0.5\end{pmatrix}
=\begin{pmatrix}-1&0.5\\0&1.5\\0&1.5\\1&2.5\end{pmatrix}\\[3mm]
%
\mathbf{X}^{(3)}&%
=\operatorname{ReLU}(\mathbf{X}^{(2)})=\begin{pmatrix}0&0.5\\0&1.5\\0&1.5\\1&2.5\end{pmatrix}\\[3mm]
%
\mathbf{X}^{(4)}&%
=\mathbf{X}^{(3)}\mathbf{W}^{(3)t}+\boldsymbol{1}\boldsymbol{b}^{(3)t}%
=\begin{pmatrix}0&0.5\\0&1.5\\0&1.5\\1&2.5\end{pmatrix}\begin{pmatrix}1&-1\\-1&1\end{pmatrix}+\begin{pmatrix}1\\1\\1\\1\end{pmatrix}(1, -1)%
%=\begin{pmatrix}-0.5&0.5\\-1.5&1.5\\-1.5&1.5\\-1.5&1.5\end{pmatrix}+\begin{pmatrix}1&-1\\1&-1\\1&-1\\1&-1\end{pmatrix}
=\begin{pmatrix}0.5&-0.5\\-0.5&0.5\\-0.5&0.5\\-0.5&0.5\end{pmatrix}\\[3mm]
%
\hat{\mathbf{Y}}=\mathbf{X}^{(5)}&%
=\boldsymbol{\mathcal{S}}(\mathbf{X}^{(4)})%
=\boldsymbol{\mathcal{S}}\begin{pmatrix}0.5&-0.5\\-0.5&0.5\\-0.5&0.5\\-0.5&0.5\end{pmatrix}%
=\begin{pmatrix}0.7311&0.2689\\0.2689&0.7311\\0.2689&0.7311\\0.2689&0.7311\end{pmatrix}\\[3mm]
%
\mathcal{L}=x^{(6)}&%
%=-\frac{1}{N}\sum_n\boldsymbol{y}_n^t\log(\hat{\boldsymbol{y}}_n)%
=-\frac{1}{N}\boldsymbol{1}^t(\mathbf{Y}\odot\log\hat{\mathbf{Y}})\,\boldsymbol{1}%
=-\frac{1}{4}(1, 1, 1, 1)\begin{pmatrix}-0.3132&0\\0&-0.3132\\0&-0.3132\\-1.3134&0\end{pmatrix}\begin{pmatrix}1\\1\end{pmatrix}%
=0.5633
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ for one data sample and generalization to batch
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}&%
=-\boldsymbol{y}^t\oslash\hat{\boldsymbol{y}}^t%
&\frac{\partial\mathcal{L}}{\partial\hat{\mathbf{Y}}}&%
=-\frac{1}{N}(\mathbf{Y}\oslash\hat{\mathbf{Y}})^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\boldsymbol{x}^{(4)}}
=(\hat{\boldsymbol{y}}-\boldsymbol{y})^t%
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\mathbf{Y}}}\frac{\partial\hat{\mathbf{Y}}}{\partial\mathbf{X}^{(4)}}%
=\frac{1}{N}(\hat{\mathbf{Y}}-\mathbf{Y})^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}\frac{\partial\boldsymbol{x}^{(4)}}{\partial\mathbf{W}^{(3)}}%
=\boldsymbol{x}^{(3)}\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial\mathbf{W}^{(3)}}%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}\frac{\partial\boldsymbol{x}^{(4)}}{\partial\boldsymbol{b}^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}\odot\boldsymbol{1}^t
&\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial\boldsymbol{b}^{(3)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}\frac{\partial\boldsymbol{x}^{(4)}}{\partial\boldsymbol{x}^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(4)}}\,\mathbf{W}^{(3)}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial\mathbf{X}^{(3)}}%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\frac{\partial\boldsymbol{x}^{(3)}}{\partial\boldsymbol{x}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\odot\operatorname{ReLU}'(\boldsymbol{x}^{(2)t})
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\frac{\partial\mathbf{X}^{(3)}}{\partial\mathbf{X}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\operatorname{ReLU}'(\mathbf{X}^{(2)t})\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\boldsymbol{x}^{(1)}\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(1)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\odot\boldsymbol{1}^t
&\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ Computation
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}&%
=\frac{1}{N}(\hat{\mathbf{Y}}-\mathbf{Y})^t%\\&%
%=\frac{1}{4}\left(\begin{pmatrix}0.7311&0.2689\\0.2689&0.7311\\0.2689&0.7311\\0.2689&0.7311\end{pmatrix}-\begin{pmatrix}1&0\\0&1\\0&1\\1&0\end{pmatrix}\right)^t\\&%
%=\frac{1}{4}\left(\begin{pmatrix}-0.2689&0.2689\\0.2689&-0.2689\\0.2689&-0.2689\\-0.7311&0.7311\end{pmatrix}\right)^t\\&%
%=\frac{1}{4}\begin{pmatrix}-0.2689&0.2689&0.2689&-0.7311\\0.2689&-0.2689&-0.2689&0.7311\end{pmatrix}\\&%
=\begin{pmatrix}-0.0672&0.0672&0.0672&-0.1828\\0.0672&-0.0672&-0.0672&0.1828\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t%
%=\begin{pmatrix}0&0&0&1\\0.5&1.5&1.5&2.5\end{pmatrix}\begin{pmatrix}-0.0672&0.0672\\0.0672&-0.0672\\0.0672&-0.0672\\-0.1828&0.1828\end{pmatrix}%
=\begin{pmatrix}-0.1828&0.1828\\-0.2890&0.2890\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(3)}}&%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t%
%=(1,1,1,1)\begin{pmatrix}-0.0672&0.0672\\0.0672&-0.0672\\0.0672&-0.0672\\-0.1828&0.1828\end{pmatrix}%
=(-0.1156,0.1156)\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}%
%=\begin{pmatrix}1&-1\\-1&1\end{pmatrix}\begin{pmatrix}-0.0672&0.0672&0.0672&-0.1828\\0.0672&-0.0672&-0.0672&0.1828\end{pmatrix}
=\begin{pmatrix}-0.1344&0.1344&0.1344&-0.3656\\0.1344&-0.1344&-0.1344&0.3656\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\operatorname{ReLU}'(\mathbf{X}^{(2)t})%
%=\begin{pmatrix}-0.1344&0.1344&0.1344&-0.3656\\0.1344&-0.1344&-0.1344&0.3656\end{pmatrix}\odot\begin{pmatrix}0&0&0&1\\1&1&1&1\end{pmatrix}\\&%
=\begin{pmatrix}0&0&0&-0.3656\\0.1344&-0.1344&-0.1344&0.3656\end{pmatrix}\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
=\begin{pmatrix}0&0&1&1\\0&1&0&1\end{pmatrix}
\begin{pmatrix}0&0.1344\\0&-0.1344\\0&-0.1344\\-0.3656&0.3656\end{pmatrix}
=\begin{pmatrix}-0.3656&0.2312\\-0.3656&0.2312\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
=(1,1,1,1)\begin{pmatrix}0&0.1344\\0&-0.1344\\0&-0.1344\\-0.3656&0.3656\end{pmatrix}
=(-0.3656,0.2312)
\end{align*}$$


<p style="page-break-after:always;"></p>


**Parameter update:** $\quad$ layer by layer, respecting the (transposed) format of the model
$$\begin{align*}
\mathbf{W}^{(3)t}&%
=\mathbf{W}^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}%
=\begin{pmatrix}1&-1\\-1&1\end{pmatrix}-0.1\begin{pmatrix}-0.1828&0.1828\\-0.2890&0.2890\end{pmatrix}
=\begin{pmatrix}1.0183&-1.0183\\-0.9711&0.9711\end{pmatrix}\\[3mm]%
%
\boldsymbol{b}^{(3)t}&=\boldsymbol{b}^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial b^{(3)}}%
=(1,-1)-0.1(-0.1156,0.1156)=(1.0116,-1.0116)\\[3mm]
%
\mathbf{W}^{(1)t}&%
=\mathbf{W}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}%
=\begin{pmatrix}1&1\\1&1\end{pmatrix}-0.1\begin{pmatrix}-0.3656&0.2312\\-0.3656&0.2312\end{pmatrix}%
=\begin{pmatrix}1.0366&0.9769\\1.0366&0.9769\end{pmatrix}\\[3mm]
%
\boldsymbol{b}^{(1)t}&%
=\boldsymbol{b}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}
=(-1, 0.5)-0.1(-0.3656,0.2312)=(-0.9634,0.4769)
\end{align*}$$


<p style="page-break-after:always;"></p>


```python
import numpy as np; np.set_printoptions(precision=4); import torch; from torch import nn
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__(); self.x2 = nn.Linear(2, 2); nn.init.ones_(self.x2.weight)
        self.x2.bias.data = torch.tensor([-1., .5], requires_grad=True)
        self.x3 = nn.ReLU(); self.x4 = nn.Linear(2, 2)
        self.x4.weight.data = torch.tensor([[1., -1.], [-1., 1]], requires_grad=True)
        self.x4.bias.data = torch.tensor([1., -1.], requires_grad=True); self.x5 = nn.Softmax(1)
    def forward(self, x):
        return self.x5(self.x4(self.x3(self.x2(x))))
model = NeuralNetwork(); # print(model)
X = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]]); print(f'X = {str(X.detach()).replace('\n','')}')
y = torch.tensor([[1., 0.], [0., 1.], [0., 1.], [1., 0.]]); print(f'y = {str(y.detach()).replace('\n','')}')
X2 = model.x2(X); print(f'X2 = {str(X2.detach()).replace('\n','')}')
X3 = model.x3(X2); print(f'X3 = {str(X3.detach()).replace('\n','')}')
X4 = model.x4(X3); print(f'X4 = {str(X4.detach()).replace('\n','')}')
X5 = model.x5(X4); print(f'X5 = {str(X5.detach()).replace('\n','')}')
loss = nn.CrossEntropyLoss()(X4, y); print(f'loss = {loss:.4f}')
model.train(); loss.backward()
torch.optim.SGD(model.parameters(), lr=1e-1).step()
print(f'W3 = {str(model.x4.weight.detach()).replace('\n','')}')
print(f'b3 = {str(model.x4.bias.detach()).replace('\n','')}')
print(f'W1 = {str(model.x2.weight.detach()).replace('\n','')}')
print(f'b1 = {str(model.x2.bias.detach()).replace('\n','')}')
```
