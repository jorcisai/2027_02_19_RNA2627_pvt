# 6 Exercise: XOR with sigmoid output

**Model:** $\quad$ hidden layer with two ReLUs and output layer with sigmoid for two classes
$$\begin{align*}
\boldsymbol{x}^{(1)}&\in\{0,1\}^2\\[3mm]
\boldsymbol{x}^{(2)}&=\mathbf{W}^{(1)}\boldsymbol{x}^{(1)}+\boldsymbol{b}^{(1)}\in\mathbb{R}^2\\[3mm]
\boldsymbol{x}^{(3)}&=\operatorname{ReLU}(\boldsymbol{x}^{(2)})\in\mathbb{R}^2\\[3mm]
x^{(4)}&=\mathbf{W}^{(3)}\boldsymbol{x}^{(3)}+b^{(3)}\in\mathbb{R}\\[3mm]
x^{(5)}&=\hat{y}=\sigma(x^{(4)})\in[0,1]\\[3mm]
x^{(6)}&=\mathcal{L}=\mathbb{H}(y,\hat{y})=-y\log(\hat{y})-(1-y)\log(1-\hat{y})
\end{align*}$$

**Current parameters:**
$$\mathbf{W}^{(1)}=\begin{pmatrix}1&1\\1&1\end{pmatrix}\quad \boldsymbol{b}^{(1)}=\begin{pmatrix}0\\-1\end{pmatrix}\quad \mathbf{W}^{(3)}=\begin{pmatrix}1&-1\end{pmatrix}\quad b^{(3)}=0$$

**Objective:** $\;$ backprop with $\quad\mathbf{X}=\begin{pmatrix}\boldsymbol{x}_1^t\\\boldsymbol{x}_2^t\\\boldsymbol{x}_3^t\\\boldsymbol{x}_4^t\end{pmatrix}%
=\begin{pmatrix}0&0\\0&1\\1&0\\1&1\end{pmatrix}\quad$ and $\quad\boldsymbol{y}=\begin{pmatrix}y_1\\y_2\\y_3\\y_4\end{pmatrix}=\begin{pmatrix}0\\1\\1\\0\end{pmatrix}$


<p style="page-break-after:always;"></p>


## Solution:

**Forward:**
$$\begin{align*}
\mathbf{X}^{(2)}&%
=\mathbf{X}^{(1)}\mathbf{W}^{(1)t}+\boldsymbol{1}\boldsymbol{b}^{(1)t}%
=\begin{pmatrix}0&0\\0&1\\1&0\\1&1\end{pmatrix}\begin{pmatrix}1&1\\1&1\end{pmatrix}+\begin{pmatrix}1\\1\\1\\1\end{pmatrix}(0, -1)%
%=\begin{pmatrix}0&0\\1&1\\1&1\\2&2\end{pmatrix}+\begin{pmatrix}0&-1\\0&-1\\0&-1\\0&-1\end{pmatrix}
=\begin{pmatrix}0&-1\\1&0\\1&0\\2&1\end{pmatrix}\\[3mm]
%
\mathbf{X}^{(3)}&%
=\operatorname{ReLU}(\mathbf{X}^{(2)})=\begin{pmatrix}0&0\\1&0\\1&0\\2&1\end{pmatrix}\\[3mm]
%
\mathbf{X}^{(4)}&%
=\mathbf{X}^{(3)}\mathbf{W}^{(3)t}+\boldsymbol{1}b^{(3)t}%
=\begin{pmatrix}0&0\\1&0\\1&0\\2&1\end{pmatrix}\begin{pmatrix}1\\-1\end{pmatrix}+\begin{pmatrix}1\\1\\1\\1\end{pmatrix}0%
=\begin{pmatrix}0\\1\\1\\1\end{pmatrix}\\[3mm]
%
\hat{\boldsymbol{y}}=\mathbf{X}^{(5)}&%
=\boldsymbol{\sigma}(\mathbf{X}^{(4)})%
=\begin{pmatrix}0.5000\\0.7310\\0.7310\\0.7310\end{pmatrix}\\[3mm]
%
\mathcal{L}=x^{(6)}&%
%=-\frac{1}{N}\sum_ny_n\log(\hat{y}_n)+(1-y_n)\log(1-\hat{y}_n)\\&%
=-\frac{1}{N}(\boldsymbol{y}^t\log(\hat{\boldsymbol{y}})+(\boldsymbol{1}-\boldsymbol{y})^t\log(\boldsymbol{1}-\hat{\boldsymbol{y}}))%\\&%
%=-\frac{1}{4}\left((0,1,1,0)\log\begin{pmatrix}0.5000\\0.7310\\0.7310\\0.7310\end{pmatrix}+(1,0,0,1)\log\begin{pmatrix}0.5000\\0.2690\\0.2690\\0.2690\end{pmatrix}\right)\\&%
=-\frac{1}{4}(2\log 0.7310+\log 0.5+\log 0.2690)
=0.6582
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ for one data sample and generalization to batch
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\hat{y}}&%
=\frac{\hat{y}-y}{\hat{y}(1-\hat{y})}%
&\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}&%
=\frac{1}{N}(\hat{\boldsymbol{y}}-\boldsymbol{y})^t\oslash(\hat{\boldsymbol{y}}\odot(\boldsymbol{1}-\hat{\boldsymbol{y}}))^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial x^{(4)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial x^{(4)}}=\hat{y}-y%
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\mathbf{X}^{(4)}}%
=\frac{1}{N}(\hat{\boldsymbol{y}}-\boldsymbol{y})^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial x^{(4)}}\frac{\partial x^{(4)}}{\partial\mathbf{W}^{(3)}}%
=\boldsymbol{x}^{(3)}\frac{\partial\mathcal{L}}{\partial x^{(4)}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial\mathbf{W}^{(3)}}%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial x^{(4)}}\frac{\partial x^{(4)}}{\partial b^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial x^{(4)}}\odot 1^t
&\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial b^{(3)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial x^{(4)}}\frac{\partial x^{(4)}}{\partial\boldsymbol{x}^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial x^{(4)}}\,\mathbf{W}^{(3)}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\frac{\partial\mathbf{X}^{(4)}}{\partial\mathbf{X}^{(3)}}%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\frac{\partial\boldsymbol{x}^{(3)}}{\partial\boldsymbol{x}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\odot\operatorname{ReLU}'(\boldsymbol{x}^{(2)t})
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\frac{\partial\mathbf{X}^{(3)}}{\partial\mathbf{X}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\operatorname{ReLU}'(\mathbf{X}^{(2)t})\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\boldsymbol{x}^{(1)}\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(1)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\odot\boldsymbol{1}^t
&\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ Computation
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}&%
=\frac{1}{N}(\hat{\boldsymbol{y}}-\boldsymbol{y})^t%\\&%
%=\frac{1}{4}\left(\begin{pmatrix}0.5000\\0.7310\\0.7310\\0.7310\end{pmatrix}-\begin{pmatrix}0\\1\\1\\0\end{pmatrix}\right)^t%
%=\frac{1}{4}\begin{pmatrix}0.5000\\-0.2690\\-0.2690\\0.7310\end{pmatrix}^t%
=(0.1250,-0.0673,-0.0673,0.1828)\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t%
%=\begin{pmatrix}0&1&1&2\\0&0&0&1\end{pmatrix}\begin{pmatrix}0.1250\\-0.0673\\-0.0673\\0.1828\end{pmatrix}
=\begin{pmatrix}0.2310\\0.1828\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}\right)^t%
%=(1,1,1,1)\begin{pmatrix}0.1250\\-0.0673\\-0.0673\\0.1828\end{pmatrix}
=0.1732\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(4)}}%
%=\begin{pmatrix}1\\-1\end{pmatrix}(0.1250,-0.0673,-0.0673,0.1828)
=\begin{pmatrix}0.1250&-0.0673&-0.0673&0.1828\\-0.1250&0.0673&0.0673&-0.1828\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\operatorname{ReLU}'(\mathbf{X}^{(2)t})%\\&%
%=\begin{pmatrix}0.1250&-0.0673&-0.0673&0.1828\\-0.1250&0.0673&0.0673&-0.1828\end{pmatrix}\odot\begin{pmatrix}0&1&1&1\\0&0&0&1\end{pmatrix}\\&%
=\begin{pmatrix}0&-0.0673&-0.0673&0.1828\\0&0&0&-0.1828\end{pmatrix}\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
=\begin{pmatrix}0&0&1&1\\0&1&0&1\end{pmatrix}\begin{pmatrix}0&0\\-0.0673&0\\-0.0673&0\\0.1828&-0.1828\end{pmatrix}
=\begin{pmatrix}0.1155&-0.1828\\0.1155&-0.1828\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
=(1,1,1,1)\begin{pmatrix}0&0\\-0.0673&0\\-0.0673&0\\0.1828&-0.1828\end{pmatrix}
=(0.0482,-0.1828)
\end{align*}$$


<p style="page-break-after:always;"></p>


**Parameter update:** $\quad$ layer by layer, respecting the (transposed) format of the model
$$\begin{align*}
\mathbf{W}^{(3)t}&%
=\mathbf{W}^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}%
=\begin{pmatrix}1\\-1\end{pmatrix}-0.1\begin{pmatrix}0.2310\\0.1828\end{pmatrix}%
=\begin{pmatrix}0.9769\\-1.0183\end{pmatrix}\\[3mm]%
%
b^{(3)t}&=b^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial b^{(3)}}%
=0-0.1\cdot 0.1732=-0.0173\\[3mm]
%
\mathbf{W}^{(1)t}&%
=\mathbf{W}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}%
=\begin{pmatrix}1&1\\1&1\end{pmatrix}-0.1\begin{pmatrix}0.1155&-0.1828\\0.1155&-0.1828\end{pmatrix}%
=\begin{pmatrix}0.9885&1.0183\\0.9885&1.0183\end{pmatrix}\\[3mm]
%
\boldsymbol{b}^{(1)t}&%
=\boldsymbol{b}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}
=(0, -1)-0.1\,(0.0482,-0.1828)=(-0.0048,-0.9817)
\end{align*}$$


<p style="page-break-after:always;"></p>


```python
import numpy as np; np.set_printoptions(precision=4); import torch; from torch import nn
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__(); self.x2 = nn.Linear(2, 2); nn.init.ones_(self.x2.weight)
        self.x2.bias.data = torch.tensor([0., -1.], requires_grad=True)
        self.x3 = nn.ReLU(); self.x4 = nn.Linear(2, 1)
        self.x4.weight.data = torch.tensor([[1., -1.]], requires_grad=True)
        nn.init.zeros_(self.x4.bias); self.x5 = nn.Sigmoid()
    def forward(self, x):
        return self.x5(self.x4(self.x3(self.x2(x))))
model = NeuralNetwork(); # print(model)
X = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]]); print(f'X = {str(X.detach()).replace('\n','')}')
y = torch.tensor([[0.], [1.], [1.], [0.]]); print(f'y = {str(y.detach()).replace('\n','')}')
X2 = model.x2(X); print(f'X2 = {str(X2.detach()).replace('\n','')}')
X3 = model.x3(X2); print(f'X3 = {str(X3.detach()).replace('\n','')}')
X4 = model.x4(X3); print(f'X4 = {str(X4.detach()).replace('\n','')}')
X5 = model.x5(X4); print(f'X5 = {str(X5.detach()).replace('\n','')}')
loss = nn.BCELoss()(X5, y); print(f'loss = {loss:.4f}')
model.train(); loss.backward()
torch.optim.SGD(model.parameters(), lr=1e-1).step()
print(f'W3 = {str(model.x4.weight.detach()).replace('\n','')}')
print(f'b3 = {str(model.x4.bias.detach()).replace('\n','')}')
print(f'W1 = {str(model.x2.weight.detach()).replace('\n','')}')
print(f'b1 = {str(model.x2.bias.detach()).replace('\n','')}')
```
