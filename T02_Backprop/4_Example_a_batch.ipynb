# 4 Example of regression with a batch

**Objective:** $\;$ train the above model with $\,N=2\,$ data,
$\;\mathbf{X}=\begin{pmatrix}\boldsymbol{x}_1^t\\\boldsymbol{x}_2^t\end{pmatrix}=\begin{pmatrix}1&1&1\\2&2&2\end{pmatrix}\;$
and $\;\boldsymbol{y}=\begin{pmatrix}y_1\\y_2\end{pmatrix}=\begin{pmatrix}1\\2\end{pmatrix}$

**Forward:** $\quad$ Generalizing data to batch is simple
$$\begin{align*}
\mathbf{X}^{(1)}&%
=\mathbf{X}=\begin{pmatrix}1&1&1\\2&2&2\end{pmatrix}\\%[3mm]
%
\mathbf{X}^{(2)}&%
=\mathbf{X}^{(1)}\mathbf{W}^{(1)t}+\boldsymbol{1}\boldsymbol{b}^{(1)t}%
=\begin{pmatrix}1&1&1\\2&2&2\end{pmatrix}\begin{pmatrix}1&1\\1&1\\1&1\end{pmatrix}+\begin{pmatrix}1\\1\end{pmatrix}(1, 1)%
%=\begin{pmatrix}3&3\\6&6\end{pmatrix}+\begin{pmatrix}1&1\\1&1\end{pmatrix}
=\begin{pmatrix}4&4\\7&7\end{pmatrix}\\[3mm]
%
\mathbf{X}^{(3)}&%
=\boldsymbol{\sigma}(\mathbf{X}^{(2)})%
=\boldsymbol{\sigma}\left(\begin{pmatrix}4&4\\7&7\end{pmatrix}\right)%
=\begin{pmatrix}\sigma(4)&\sigma(4)\\\sigma(7)&\sigma(7)\end{pmatrix}
=\begin{pmatrix}0.9820&0.9820\\0.9991&0.9991\end{pmatrix}\\[3mm]
%
\hat{\boldsymbol{y}}=\mathbf{X}^{(4)}&%
=\mathbf{X}^{(3)}\mathbf{W}^{(3)t}+\boldsymbol{1}b^{(3)t}
=\begin{pmatrix}0.9820&0.9820\\0.9991&0.9991\end{pmatrix}\begin{pmatrix}1\\1\end{pmatrix}+\begin{pmatrix}1\\1\end{pmatrix}1
=\begin{pmatrix}2.9640\\2.9982\end{pmatrix}\\[3mm]
%
\boldsymbol{y}-\hat{\boldsymbol{y}}&
=\begin{pmatrix}1\\2\end{pmatrix}-\begin{pmatrix}2.9640\\2.9982\end{pmatrix}=\begin{pmatrix}-1.9640\\-0.9982\end{pmatrix}\\[3mm]
%
\mathcal{L}=x^{(5)}&%
=\frac{1}{N}\sum_n(y_n-\hat{y}_n)^2=\frac{1}{N}(\boldsymbol{y}-\hat{\boldsymbol{y}})^t(\boldsymbol{y}-\hat{\boldsymbol{y}})=2.4268
\end{align*}$$

**Weight matrices and transposed bias vectors:** $\quad$ greater efficiency in forward (and inference)


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ generalization of data to batch
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\hat{y}}&=2(\hat{y}-y)%
&\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}&%
=\frac{2}{N}(\hat{\boldsymbol{y}}-\boldsymbol{y})^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial\mathbf{W}^{(3)}}%
=\boldsymbol{x}^{(3)}\frac{\partial\mathcal{L}}{\partial\hat{y}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\mathbf{W}^{(3)}}%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial b^{(3)}}%
&\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial b^{(3)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial\boldsymbol{x}^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\,\mathbf{W}^{(3)}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\mathbf{X}^{(3)}}%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\frac{\partial\boldsymbol{x}^{(3)}}{\partial\boldsymbol{x}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\odot\boldsymbol{\sigma}'(\boldsymbol{x}^{(2)t})
&\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\frac{\partial\mathbf{X}^{(3)}}{\partial\mathbf{X}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\boldsymbol{\sigma}'(\mathbf{X}^{(2)t})\\[3mm]%
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\boldsymbol{x}^{(1)}\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}%
&\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\odot\boldsymbol{1}^t
&\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ Computation
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}&%
=\frac{2}{N}(\hat{\boldsymbol{y}}-\boldsymbol{y})^t=(1.9640, 0.9982)\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
%=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\mathbf{W}^{(3)}}%
=\mathbf{X}^{(3)t}\left(\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\right)^t%
=\begin{pmatrix}0.9820&0.9991\\0.9820&0.9991\end{pmatrix}\begin{pmatrix}1.9640\\0.9982\end{pmatrix}
=\begin{pmatrix}2.9259\\2.9259\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
%=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial b^{(3)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\right)^t%
=(1, 1)\begin{pmatrix}1.9640\\0.9982\end{pmatrix}=2.9622\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}&%
%=\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}\frac{\partial\hat{\boldsymbol{y}}}{\partial\mathbf{X}^{(3)}}%
=\mathbf{W}^{(3)t}\frac{\partial\mathcal{L}}{\partial\hat{\boldsymbol{y}}}%
=\begin{pmatrix}1\\1\end{pmatrix}(1.9640, 0.9982)=\begin{pmatrix}1.9640&0.9982\\1.9640&0.9982\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}&%
%=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\frac{\partial\mathbf{X}^{(3)}}{\partial\mathbf{X}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(3)}}\odot\boldsymbol{\sigma}'(\mathbf{X}^{(2)t})%
=\begin{pmatrix}1.9640&0.9982\\1.9640&0.9982\end{pmatrix}\odot\begin{pmatrix}\sigma'(4)&\sigma'(7)\\\sigma'(4)&\sigma'(7)\end{pmatrix}%
=\begin{pmatrix}0.0347&0.0009\\0.0347&0.0009\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
%=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\mathbf{X}^{(1)t}\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t%
=\begin{pmatrix}1&2\\1&2\\1&2\end{pmatrix}\begin{pmatrix}0.0347&0.0347\\0.0009&0.0009\end{pmatrix}%
=\begin{pmatrix}0.0365&0.0365\\0.0365&0.0365\\0.0365&0.0365\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
%=\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\frac{\partial\mathbf{X}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\boldsymbol{1}^t\left(\frac{\partial\mathcal{L}}{\partial\mathbf{X}^{(2)}}\right)^t%
=(1, 1)\begin{pmatrix}0.0347&0.0347\\0.0009&0.0009\end{pmatrix}%
=(0.0356, 0.0356)
\end{align*}$$


<p style="page-break-after:always;"></p>


**Parameter update:** $\quad$ layer by layer, respecting the (transposed) format of the model
$$\begin{align*}
\mathbf{W}^{(3)t}&%
=\mathbf{W}^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}%
=\begin{pmatrix}1\\1\end{pmatrix}-0.1\begin{pmatrix}2.9259\\2.9259\end{pmatrix}%
=\begin{pmatrix}0.7074\\0.7074\end{pmatrix}\\[3mm]%
%
b^{(3)t}&=b^{(3)t}-\eta\,\frac{\partial\mathcal{L}}{\partial b^{(3)}}=1-0.1\cdot 2.9622=0.7038\\[3mm]
%
\mathbf{W}^{(1)t}&%
=\mathbf{W}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}%
=\begin{pmatrix}1&1\\1&1\\1&1\end{pmatrix}-0.1\begin{pmatrix}0.0365&0.0365\\0.0365&0.0365\\0.0365&0.0365\end{pmatrix}%
=\begin{pmatrix}0.9964&0.9964\\0.9964&0.9964\\0.9964&0.9964\end{pmatrix}\\[3mm]
%
\boldsymbol{b}^{(1)t}&%
=\boldsymbol{b}^{(1)t}-\eta\,\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}=(1, 1)-0.1(0.0356, 0.0356)=(0.9964, 0.9964)
\end{align*}$$


<p style="page-break-after:always;"></p>


```python
import numpy as np; np.set_printoptions(precision=4); import torch; from torch import nn
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.x2 = nn.Linear(3, 2); nn.init.ones_(self.x2.weight); nn.init.ones_(self.x2.bias)
        self.x3 = nn.Sigmoid()
        self.x4 = nn.Linear(2, 1); nn.init.ones_(self.x4.weight); nn.init.ones_(self.x4.bias)
    def forward(self, x):
        return self.x4(self.x3(self.x2(x)))
model = NeuralNetwork(); print(model)
X = torch.tensor([[1., 1., 1.], [2., 2., 2.]]); print(f'X = {str(X.detach()).replace('\n','')}')
y = torch.tensor([[1.], [2.]]); print(f'y = {str(y.detach()).replace('\n','')}')
X2 = model.x2(X); print(f'X2 = {str(X2.detach()).replace('\n','')}')
X3 = model.x3(X2); print(f'X3 = {str(X3.detach()).replace('\n','')}')
X4 = model.x4(X3); print(f'X4 = {str(X4.detach()).replace('\n','')}')
loss = nn.MSELoss()(X4, y); print(f'loss = {loss:.4f}')
model.train(); loss.backward()
torch.optim.SGD(model.parameters(), lr=1e-1).step()
print(f'W3 = {model.x4.weight.detach()}')
print(f'b3 = {model.x4.bias.detach()}')
print(f'W1 = {model.x2.weight.detach()}')
print(f'b1 = {model.x2.bias.detach()}')
```
