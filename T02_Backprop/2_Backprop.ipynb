# 2 Backprop algorithm

**Backprop Algorithm (Backpropagation):**
* Calculates the loss gradient with respect to the parameters of each layer
* The calculated gradient is passed to an optimization algorithm
* Origin: 1982; Paul Werbos applied it to MLPs in the way that has become standard
* Popularization: Rumelhart, Hinton, and Williams popularized it in a 1986 Nature article

**Input:** $\;$ MLP with $\,L-1\,$ layers, each with its $\,\boldsymbol{\theta}^{(\ell)}\,$ ($\emptyset\,$ if it doesn't have one), and loss with respect to $(\boldsymbol{x},\boldsymbol{y})$ in "layer" $L$
$$\boldsymbol{x}^{(1)}=\boldsymbol{x}\,\to\,%
\boldsymbol{x}^{(2)}=\boldsymbol{f}^{(1)}(\boldsymbol{x}^{(1)}, \boldsymbol{\theta}^{(1)})\,\overset{\cdots}{\to}\,%
\boldsymbol{x}^{(L)}=\boldsymbol{f}^{(L-1)}(\boldsymbol{x}^{(L-1)}, \boldsymbol{\theta}^{(L-1)})\,\to\,%
x^{(L+1)}=\boldsymbol{f}^{(L)}(\boldsymbol{x}^{(L)},\boldsymbol{y})=\mathcal{L}$$

**Method:** $\quad$ loop that executes two steps in each iteration

* **Forward:** $\quad$ traverses the model forward to find the output of each layer from an input $\boldsymbol{x}$

* $\boldsymbol{x}^{(1)}=\boldsymbol{x}$

* **for** $\;\ell=1:L:\quad\boldsymbol{x}^{(\ell+1)}=\boldsymbol{f}^{(\ell)}(\boldsymbol{x}^{(\ell)},\boldsymbol{\theta}^{(\ell)})$

* **Backward:** $\quad$ traverses the model backward to find the gradient of the loss with respect to the parameters

* $u^{(L+1)}=1$

* **for** $\;\ell=L:1:\quad\boldsymbol{g}^{(\ell)}=\boldsymbol{u}^{(\ell+1)t}\dfrac{\partial\boldsymbol{x}^{(\ell+1)}}{\partial\boldsymbol{\theta}^{( \ell)}}\qquad\boldsymbol{u}^{(\ell)t}=\boldsymbol{u}^{(\ell+1)t}\dfrac{\partial\boldsymbol{x}^{(\ell+1)}}{\partial\boldsymbol{x}^{(\ell)}}$
