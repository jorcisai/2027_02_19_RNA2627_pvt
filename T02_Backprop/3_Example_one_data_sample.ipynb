$$\boldsymbol{x}^{(1)}\in\mathbb{R}^3\,\to\,%
\boldsymbol{x}^{(2)}=\mathbf{W}^{(1)}\boldsymbol{x}^{(1)}+\boldsymbol{b}^{(1)}\in\mathbb{R}^2\,\to\,%
\boldsymbol{x}^{(3)}=\boldsymbol{\sigma}(\boldsymbol{x}^{(2)})\in\mathbb{R}^2\,\to\,%
x^{(4)}=\mathbf{W}^{(3)}\boldsymbol{x}^{(3)}+b^{(3)}\in\mathbb{R}$$

**Objective:** $\;$ given $\,(\boldsymbol{x}, y)=(\boldsymbol{1}, 1),\,$ minimize the quadratic loss by one iteration of GD with $\,\eta=0.1$


<p style="page-break-after:always;"></p>


**Current parameters:** $\quad$ Let's assume that $\,\boldsymbol{\theta}=\boldsymbol{1},\,$ or, following the model format,
$$\mathbf{W}^{(1)}=\begin{pmatrix}1&1&1\\1&1&1\end{pmatrix}\qquad\boldsymbol{b}^{(1)}=\begin{pmatrix}1\\1\end{pmatrix}%
\qquad\mathbf{W}^{(3)}=(1, 1)\qquad b^{(3)}=1$$

**Loss:** To find the gradient of the quadratic loss, we consider the loss as an additional "layer":
$$\boldsymbol{x}^{(1)}=\boldsymbol{x}\,\to\,\boldsymbol{x}^{(2)}\,\to\,\boldsymbol{x}^{(3)}\,\to\,x^{(4)}=\hat{y}\,\to\,x^{(5)}=\mathcal{L}=(y-\hat{y})^2$$

**Forward:** $\quad$ recorremos el modelo hacia adelante para hallar la salida de cada capa a partir de la entrada
$$\begin{align*}
\boldsymbol{x}^{(1)}&%
=\begin{pmatrix}1\\1\\1\end{pmatrix}\\[3mm]
%
\boldsymbol{x}^{(2)}&%
=\mathbf{W}^{(1)}\boldsymbol{x}^{(1)}+\boldsymbol{b}^{(1)}
=\begin{pmatrix}1&1&1\\1&1&1\end{pmatrix}\begin{pmatrix}1\\1\\1\end{pmatrix}+\begin{pmatrix}1\\1\end{pmatrix}%
=\begin{pmatrix}3\\3\end{pmatrix}+\begin{pmatrix}1\\1\end{pmatrix}%
=\begin{pmatrix}4\\4\end{pmatrix}\\[3mm]
%
\boldsymbol{x}^{(3)}&%
=\boldsymbol{\sigma}(\boldsymbol{x}^{(2)})%
=\boldsymbol{\sigma}\left(\begin{pmatrix}4\\4\end{pmatrix}\right)%
=\begin{pmatrix}\sigma(4)\\\sigma(4)\end{pmatrix}%
=\begin{pmatrix}0.982\\0.982\end{pmatrix}\\[3mm]
%
\hat{y}=x^{(4)}&%
=\mathbf{W}^{(3)}\boldsymbol{x}^{(3)}+b^{(3)}%
=(1,1)\begin{pmatrix}0.982\\0.982\end{pmatrix}+1%
=2.9640\\[3mm]
%
\mathcal{L}=x^{(5)}&=(y-\hat{y})^2=(1-2.9640)^2=3.8574
\end{align*}$$


<p style="page-break-after:always;"></p>


**Backward:** $\quad$ We traverse the model backwards to find the gradient of the loss with respect to the parameters
$$\begin{align*}
\frac{\partial\mathcal{L}}{\partial\hat{y}}&=2(\hat{y}-y)=3.9280\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial\mathbf{W}^{(3)}}%
=\boldsymbol{x}^{(3)}\frac{\partial\mathcal{L}}{\partial\hat{y}}=\begin{pmatrix}3.8574\\3.8574\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial b^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial b^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\cdot 1=3.9280\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}&%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\frac{\partial\hat{y}}{\partial\boldsymbol{x}^{(3)}}%
=\frac{\partial\mathcal{L}}{\partial\hat{y}}\,\mathbf{W}^{(3)}%
=(3.9280, 3.9280)\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\frac{\partial\boldsymbol{x}^{(3)}}{\partial\boldsymbol{x}^{(2)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(3)}}\odot\boldsymbol{\sigma}'(\boldsymbol{x}^{(2)t})%
=(0.0694, 0.0694)\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\mathbf{W}^{(1)}}%
=\boldsymbol{x}^{(1)}\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}%
=\begin{pmatrix}0.0694&0.0694\\0.0694&0.0694\\0.0694&0.0694\end{pmatrix}\\[3mm]
%
\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}&%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\frac{\partial\boldsymbol{x}^{(2)}}{\partial\boldsymbol{b}^{(1)}}%
=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}\odot\boldsymbol{1}^t=\frac{\partial\mathcal{L}}{\partial\boldsymbol{x}^{(2)}}=(0.0694, 0.0694)%
\end{align*}$$


<p style="page-break-after:always;"></p>


**Parameter update:** $\quad$ We do it layer by layer, respecting the model format
$$\begin{align*}
\mathbf{W}^{(3)}&%
=\mathbf{W}^{(3)}-\eta\left(\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(3)}}\right)^t\\%
&=(1, 1)-0.1\,(3.8574, 3.8574)=(0.6143, 0.6143)\\[3mm]%
%
b^{(3)}&%
=b^{(3)}-\eta\left(\frac{\partial\mathcal{L}}{\partial b^{(3)}}\right)^t\\%
&=1-0.1\cdot 3.9280=0.6072\\[3mm]
%
\mathbf{W}^{(1)}&%
=\mathbf{W}^{(1)}-\eta\left(\frac{\partial\mathcal{L}}{\partial\mathbf{W}^{(1)}}\right)^t\\%
&=\begin{pmatrix}1&1&1\\1&1&1\end{pmatrix}-0.1\begin{pmatrix}0.0694&0.0694&0.0694\\0.0694&0.0694&0.0694\end{pmatrix}
=\begin{pmatrix}0.9931&0.9931&0.9931\\0.9931&0.9931&0.9931\end{pmatrix}\\[3mm]
%
\boldsymbol{b}^{(1)}&%
=\boldsymbol{b}^{(1)}-\eta\left(\frac{\partial\mathcal{L}}{\partial\boldsymbol{b}^{(1)}}\right)^t\\%
&=\begin{pmatrix}1\\1\end{pmatrix}-0.1\begin{pmatrix}0.0694\\0.0694\end{pmatrix}%
=\begin{pmatrix}0.9931\\0.9931\end{pmatrix}
\end{align*}$$


<p style="page-break-after:always;"></p>


```python
import numpy as np; np.set_printoptions(precision=4); import torch; from torch import nn
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.x2 = nn.Linear(3, 2); nn.init.ones_(self.x2.weight); nn.init.ones_(self.x2.bias)
        self.x3 = nn.Sigmoid()
        self.x4 = nn.Linear(2, 1); nn.init.ones_(self.x4.weight); nn.init.ones_(self.x4.bias)
    def forward(self, x):
        return self.x4(self.x3(self.x2(x)))
model = NeuralNetwork(); print(model)
x = torch.ones(1, 3); print(f'x = {x.detach()}')
y = torch.ones(1, 1); print(f'y = {y.detach()}')
x2 = model.x2(x); print(f'x2 = {x2.detach()}')
x3 = model.x3(x2); print(f'x3 = {x3.detach()}')
x4 = model.x4(x3); print(f'x4 = {x4.detach()}')
loss = nn.MSELoss()(x4, y); print(f'loss = {loss:.4f}')
model.train(); loss.backward()
torch.optim.SGD(model.parameters(), lr=1e-1).step()
print(f'W3 = {model.x4.weight.detach()}')
print(f'b3 = {model.x4.bias.detach()}')
print(f'W1 = {model.x2.weight.detach()}')
print(f'b1 = {model.x2.bias.detach()}')
```
